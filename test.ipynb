```python
# write a function that extracts the answer from the answer column of the df and stores it in a new column called extracted_answer and the answer is the last number in the whole answer string.
import pandas as pd
import re

def extract_last_number(df , column_name , store_name):
    """
    Extracts the last number from the 'answer' column of the DataFrame
    and stores it in a new column called 'extracted_answer'.

    Parameters:
        df (pd.DataFrame): Input DataFrame with an 'answer' column.

    Returns:
        pd.DataFrame: Modified DataFrame with the new 'extracted_answer' column.
    """
    def get_last_number(s):
        numbers = re.findall(r'\d+', str(s))
        return int(numbers[-1]) if numbers else None

    df[store_name] = df[column_name].apply(get_last_number)
    return df
```

```python
import pandas as pd

def compare_answers(df):
    """
    Compares 'extracted_answer' and 'pred_answer_value' row-wise,
    assigns 1 if equal, else 0, and calculates accuracy.

    Parameters:
        df (pd.DataFrame): DataFrame with 'extracted_answer' and 'pred_answer_value' columns.

    Returns:
        pd.DataFrame: Updated DataFrame with 'match_with_answer' column.
        float: Accuracy score (0 to 1).
    """
    def is_match(extracted, predicted):
        try:
            return int(extracted) == int(predicted)
        except:
            return 0

    df['match_with_answer'] = df.apply(
        lambda row: is_match(row['extracted_answer'], row['pred_answer_value']),
        axis=1
    )

    accuracy = df['match_with_answer'].mean()  # Mean gives accuracy
    return df, accuracy
```

```python
df, accuracy = compare_answers(df)

print("Accuracy:", accuracy)
print(df.head())
```

```python
# separate those datapoint in a new dataframe where df_1["match_with_answer"] == 0
df_1_match = df_1[df_1["match_with_answer"] == 1]

df_1_match.head()
```

```python
import random 

def permute_questions(df):
    new_rows = []

    for _, row in df.iterrows():
        question = row['question']
        answer = row['answer']
        
        # Split the question into sentences
        sentences = question.split('.')
        if len(sentences) < 3:  # Ensure there are at least 3 sentences
            continue
        
        # Separate the first and last sentence
        first_sentence = sentences[0]
        last_sentence = sentences[-1]
        middle_sentences = sentences[1:-1]
        
        # Generate 3 permutations of the middle sentences
        for _ in range(3):
            shuffled_middle = middle_sentences[:]
            random.shuffle(shuffled_middle)
            permuted_question = f"{first_sentence} {' '.join(shuffled_middle)} {last_sentence}."
            new_rows.append({'question': permuted_question, 'answer': answer})
    
    # Create a new DataFrame with the permuted questions
    new_df = pd.DataFrame(new_rows)
    return new_df
```

```python
new_df = permute_questions(df_1_match)
```
